# Subplotting

```python
from sklearn import datasets
# The 1st column is sepal length and the 2nd column is sepal width
sepals = datasets.load_iris().data[:, :2]
df = pd.read_csv('data/playgolf.csv', delimiter='|', parse_dates=['Date'])
```

```python
plt.figure(figsize=(8,8))

plt.subplot(221)
plt.scatter(sepals[:,0], sepals[:,1])
plt.xlabel('length')
plt.ylabel('width')
plt.title('Flowers')

plt.subplot(222)
df.plot(kind='scatter', x='Temperature', y='Humidity', ax=plt.gca(), \
        c=df['Result'].map({"Don't Play":'r', 'Play':'b'}), s=(df['Humidity']-50)*10)
plt.title('Weather')

plt.subplot(223)
plt.scatter(foo, bar)
plt.title('Random numbers')
plt.xlabel('foo')
plt.ylabel('bar')

plt.subplot(224)
x = np.linspace(0,0.3,10000)
plt.plot(x, np.sin(1./x))
plt.title('Cool Function')
plt.xlabel('x')
plt.ylabel('y')

plt.tight_layout()
```
